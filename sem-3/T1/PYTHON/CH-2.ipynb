```python
# WAP to encrypt to given numberic password such that the encryption shows all individual values higher by 1 count
def encrypt_number(num):
    result = 0
    place = 1
    
    while num > 0:
        digit = num % 10  
        new_digit = (digit + 1) % 10
        result += new_digit * place
        num //= 10
        place *= 10
    
    return result

# Example
password = int(input("Enter password"))
encrypted = encrypt_number(password)
print("Original:", password)
print("Encrypted:", encrypted)
```
