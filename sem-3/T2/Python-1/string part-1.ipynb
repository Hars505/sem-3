#WAP to fetch any word from user the pogram should display the positive and negative index of each number

```python
word=str(input("Enter any word : "))
for i in range(len(word)):
    print(f" The Postive index of {word[i]} is {i} and negative is {i-len(word)}")
```

```python
# WAP To reverse any given word without usoing slice operator
word=str(input("Enter any word : "))
print("Reverse word is : ",end=" ")
for i in range(len(word)):
    print(f"{word[len(word)-i-1]}",end='')
```
