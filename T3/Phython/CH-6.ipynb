```python
# WAP TO find total number if vowels total characters total words from the given file
F=open("alice_in_wonderland.txt")
content=F.read()
count=0
count_vowels=0
count_consonants=0
count_punct=0
for i in content:
    if i in ('a','e','i','o','u','A','E','I','O','U'):
        count_vowels+=1
    elif i.isalpha()==True or not ('a','e','i','o','u','A','E','I','O','U'):
        count_consonants+=1
    else:
        count_punct+=1
F.seek(0)
print(f"{len(F.read().split())} is total number of words")
print(f"{count_vowels} is total number of vowels")
print(f"{count_consonants} is total number of consonants")
print(f"{count_punct} is total numbers of space and punctuation")
```

## WAP USING A file which consists of multiple statements find all the words from that file that can be made using set of characters given by user

```python
F1=open("file15.txt")
LT=set(eval(input("Enter letters (in list): ").upper()))
A=[]
for line in F1:
    words = line.upper().split()
    for word in words:
        if LT.issubset(set(word)):
            A.append(word)   
if len(A)==0:
    print("There is no such word in file that has all the letters that you input")
else:
    for i in A:print(i)
```
